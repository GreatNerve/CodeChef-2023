```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy
```

```python
df = pd.read_csv('sales_data.csv')
```

# create a line plot showing the revenue over ID demonstrating how revenue changes as ID increases

```python
plt.plot(df['ID'], df['Revenue'])
plt.title('I=Revenue')
plt.xlabel('ID')
plt.ylabel('Revenue')
plt.show()
```

# the scatter plot should display the relationship between quantity and revenue. (Each point represents a transaction)

```python
plt.scatter(df['Quantity'], df['Revenue'])
plt.title('Quantity VS Revenue')
plt.xlabel('Quantity')
plt.ylabel('Revenue')
plt.show()
```

# the code should generate a bar chart depicting the total revenue for each product category.

```python
s = set(df['Product'])
pair = zip(s,[0 for i in range(len(s))])
dic = dict(pair)
```

```python
for i in range(len(df)):
    dic[df.iloc[i]['Product']]+=df.iloc[i]['Revenue']
```

```python
plt.bar(dic.keys(), dic.values())
plt.title('Product VS Revenue')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```
